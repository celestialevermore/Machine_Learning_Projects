##### Q. 연령대에 따라 이름을 달리하는 get_category함수를 만들어서 apply로 넘겨서 Age 컬럼을 다변화해보세요.

```python
def get_category(age):
    res = ''
    if age<=5 : res='Baby'
    elif age<=12: res = 'Teenager'
    elif age<=18: res = 'Student'
    elif age<=25: res = 'Young Adult'
    elif age<=35: res = 'Adult'
    else: res = 'Elderly'
        
    return res

titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : get_category(x))

titanic_df[['Age','Age_cat']].head(10)
```
